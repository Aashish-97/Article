```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Load the dataset (Ensure the file is uploaded to Colab)
file_path = "retail_India.xlsx"  # Change this if the file has a different name
xls = pd.ExcelFile(file_path)

# Load the raw data sheet
df = pd.read_excel(xls, sheet_name="rawdata")

# Drop rows where profit or freight mode is missing
df = df.dropna(subset=["Profit", "Freight Mode"])

# Perform One-Way ANOVA
freight_modes = df["Freight Mode"].unique()
profit_data = [df[df["Freight Mode"] == mode]["Profit"] for mode in freight_modes]

anova_result = stats.f_oneway(*profit_data)

# Print ANOVA results
print("One-Way ANOVA Results:")
print(f"F-statistic: {anova_result.statistic:.4f}")
print(f"P-value: {anova_result.pvalue:.4f}")

# Interpret results
alpha = 0.05
if anova_result.pvalue < alpha:
    print("Result: Significant difference in profit means across freight modes.")
else:
    print("Result: No significant difference in profit means across freight modes.")

# Visualization - Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x="Freight Mode", y="Profit", data=df, palette="Set2")
plt.title("Profit Distribution Across Freight Modes")
plt.xlabel("Freight Mode")
plt.ylabel("Profit")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
# Import required libraries
import pandas as pd
import scipy.stats as stats

# Load dataset
file_path = "retail_India.xlsx"  # Ensure file is uploaded in Google Colab
xls = pd.ExcelFile(file_path)
df = pd.read_excel(xls, sheet_name="rawdata")

# Drop missing values in Profit and Freight Mode
df = df.dropna(subset=["Profit", "Freight Mode"])

# Compute descriptive statistics for each Freight Mode
summary_stats = df.groupby("Freight Mode")["Profit"].describe()
print("Summary Statistics of Profit by Freight Mode:\n")
print(summary_stats)

# Perform One-Way ANOVA
freight_modes = df["Freight Mode"].unique()
profit_data = [df[df["Freight Mode"] == mode]["Profit"] for mode in freight_modes]

anova_result = stats.f_oneway(*profit_data)

# Print ANOVA results
print("\nOne-Way ANOVA Results:")
print(f"F-statistic: {anova_result.statistic:.4f}")
print(f"P-value: {anova_result.pvalue:.4f}")

# Interpret results
alpha = 0.05
if anova_result.pvalue < alpha:
    print("\nInterpretation: The p-value is less than 0.05, indicating a significant difference in profit means across freight modes.")
    print("This means that the choice of Freight Mode affects profitability.")
else:
    print("\nInterpretation: The p-value is greater than 0.05, suggesting that there is NO significant difference in mean profits across freight modes.")
    print("This means the choice of Freight Mode does NOT significantly impact profitability.")

# Interpretation of Boxplot
print("\nBoxplot Interpretation:")
print("- The median profit for each freight mode can be seen as the center line in each box.")
print("- The spread of the data (Interquartile Range, IQR) varies, indicating different variability in profits.")
print("- Outliers (dots outside the whiskers) suggest occasional extreme profit/loss cases.")
print("- If the ANOVA test was significant, it confirms that the visible profit differences in the boxplot are statistically meaningful.")
print("- If the ANOVA test was NOT significant, the variations in profit across freight modes are likely due to random fluctuations rather than a real impact.")

```
